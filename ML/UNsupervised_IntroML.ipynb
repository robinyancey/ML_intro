## **1.0 Unsupervised Learning vs. Supervised Learning**

At the highest level, machine learning models can be categorized by the type of data they learn from. The most common distinction is between **Supervised** and **Unsupervised** learning.

**Supervised Learning:** In supervised learning, the model learns from data that is **labeled**. This means for each data point, there is a known outcome or 'ground truth'. The model's goal is to learn a mapping function that can predict the output for new, unseen data.
  * **Analogy:** Learning with a teacher. The teacher provides examples (questions) and the correct answers.

**Unsupervised Learning:** In unsupervised learning, the model works with data that is **unlabeled**. The goal is to find hidden patterns, structures, or relationships within the data itself, without any pre-existing outcomes to guide the process.
  * **Analogy:** Learning without a teacher. You are given a box of mixed items and asked to sort them into groups based on their inherent similarities.

### **Comparison and Applications**

| **Input Data** | Labeled data (features and target outcomes)         | Unlabeled data (features only)                      |

## **1.1 Categories of Unsupervised Learning**

Unsupervised learning is not a single algorithm but a class of them. The main categories are:

1.  **Clustering:** The task of grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar to each other than to those in other groups.

3.  **Dimensionality Reduction:** The process of reducing the number of random variables (features) under consideration, by obtaining a set of principal variables. It is often used to simplify data for easier visualization and analysis.
    * *Example:* Compressing high-dimensional genetic data into a few key components to visualize population structures.

## **1.2 Problems Encountered in Unsupervised Learning**

While powerful, unsupervised learning presents unique challenges:

-   **No Ground Truth:** Since there are no labels, it's difficult to evaluate the performance of an algorithm. Is one clustering result 'better' than another? The answer is often subjective.
-   **Interpretation:** The results, such as clusters or reduced dimensions, require domain knowledge to be interpreted and validated. The algorithm finds a pattern, but a human must decide if it's meaningful.
-   **Parameter Tuning:** Many algorithms require you to specify the number of clusters (like 'k' in k-Means) or other parameters without a clear 'correct' value.
-   **The Curse of Dimensionality:** Algorithms can perform poorly on high-dimensional data, as the concept of distance becomes less meaningful.

## **1.3 Data Processing and Scaling**

Many unsupervised algorithms, especially clustering methods like K-Means, are **distance-based**. This means they rely on calculating the distance between data points to determine similarity.

If features are on vastly different scales (e.g., one feature is 'age' from 20-70, and another is 'income' from 50,000-200,000), the feature with the larger scale will dominate the distance calculation. This can lead to biased and incorrect results.

**Scaling** ensures that all features contribute equally to the distance metric. Common scaling techniques include:

-   **StandardScaler:** Rescales data to have a mean of 0 and a standard deviation of 1.
-   **MinMaxScaler:** Rescales data to a fixed range, usually 0 to 1.

```python
# Import necessary libraries for data manipulation, scaling, and visualization
import numpy as np  # For numerical operations and random number generation
import pandas as pd  # For creating and working with DataFrames (structured data)
from sklearn.preprocessing import StandardScaler, MinMaxScaler  # Tools for scaling/normalizing data
import matplotlib.pyplot as plt  # For creating plots and charts
import seaborn as sns  # For enhanced statistical visualizations

# Generate some sample data with different scales
# Setting a seed ensures we get the same random numbers each time we run this code
np.random.seed(42)

# Create a DataFrame with two columns that have very different scales:
# - income: ranges from $25,000 to $150,000 (large numbers)
# - age: ranges from 18 to 65 years (small numbers)
# This difference in scales is what we'll demonstrate scaling can fix
data = pd.DataFrame({
    'income': np.random.randint(25000, 150000, 100),  # Generate 100 random income values
    'age': np.random.randint(18, 65, 100)  # Generate 100 random age values
})

# Apply StandardScaler
# StandardScaler transforms data so it has mean=0 and standard deviation=1
# This puts all features on the same scale regardless of their original range
scaler = StandardScaler()

# fit_transform() does two things:
# 1. fit(): learns the mean and std of the data
# 2. transform(): applies the scaling formula: (value - mean) / std
scaled_data = scaler.fit_transform(data)

# Convert the scaled numpy array back to a DataFrame for easier handling
# Keep the same column names as the original data
scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

# Visualize the effect of scaling
# Create a figure with 2 subplots side by side to compare before/after
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Left plot: Show original data
ax1.set_title('Before Scaling')
# Create a scatter plot where each point represents one person's income vs age
sns.scatterplot(x='income', y='age', data=data, ax=ax1)
# Set consistent axis limits to make comparison easier
ax1.set_xlim(0, 160000)  # Income axis from 0 to 160k
ax1.set_ylim(0, 70)      # Age axis from 0 to 70

# Right plot: Show scaled data
ax2.set_title('After StandardScaler')
# Same type of scatter plot but with scaled data
sns.scatterplot(x='income', y='age', data=scaled_df, ax=ax2)
# After scaling, values typically range from about -3 to +3
ax2.set_xlim(-3, 3)  # Scaled values usually fall within this range
ax2.set_ylim(-3, 3)

# Adjust spacing between plots so they don't overlap
plt.tight_layout()
# Display the plots
plt.show()

# Print summary statistics to show the numerical effect of scaling
# Original data: income has mean ~87k, age has mean ~41
# Notice how different the scales are
print("Original Data Summary:\n", data.describe().loc[['mean', 'std']])

# Scaled data: both features now have mean ≈ 0 and std ≈ 1
# This is exactly what StandardScaler is designed to do
print("\nScaled Data Summary:\n", scaled_df.describe().loc[['mean', 'std']])
```

## **1.4 Preprocessing Methods (pandas)**

Before scaling, data often needs cleaning. Real-world datasets are messy. Common issues include missing values and categorical data.

-   **Handling Missing Values:** Most ML algorithms cannot handle missing values (`NaN`). They must be either removed or imputed (filled in with a value like the mean, median, or a constant).
-   **Encoding Categorical Variables:** Algorithms work with numbers, not text. Categorical features (like 'Country' or 'Gender') must be converted into a numerical format. A common method is one-hot encoding, which creates a new binary column for each category.

```python
# Create a sample DataFrame with messy data
# This DataFrame intentionally contains common real-world data problems:
# - Missing values (NaN = "Not a Number")
# - Categorical data mixed with numerical data
# This simulates what you'd often encounter when working with real datasets
messy_data = pd.DataFrame({
    'age': [25, 30, np.nan, 45, 22],  # Notice the np.nan (missing value) in position 2
    'city': ['New York', 'London', 'Paris', 'New York', 'Tokyo'],  # Text/categorical data
    'salary': [50000, np.nan, 70000, 80000, 45000]  # Another missing value in position 1
})

# Display the original messy data so we can see what we're starting with
print("Original Messy Data:")
print(messy_data)

# --- Preprocessing Steps ---
# Create a copy of the original data to work with
# This preserves the original data in case we need to reference it later
# It's a good practice to never modify original data directly
cleaned_data = messy_data.copy()

# 1. Impute missing 'age' with the median
# "Impute" means to fill in missing values with estimated values
# We use median instead of mean because median is less affected by outliers
# For example: ages 20, 25, 30, 100 → median=27.5, mean=43.75
# The median (27.5) is more representative of typical age in this case
age_median = cleaned_data['age'].median()  # Calculate median of non-missing age values
cleaned_data['age'].fillna(age_median, inplace=True)  # Replace NaN with median
# inplace=True means modify the DataFrame directly instead of returning a new one

# 2. Impute missing 'salary' with the mean
# For salary, we use mean instead of median as an example of different strategies
# In practice, you'd choose based on your data's distribution and business context
# Mean works well when data is roughly normally distributed without extreme outliers
salary_mean = cleaned_data['salary'].mean()  # Calculate average of non-missing salary values
cleaned_data['salary'].fillna(salary_mean, inplace=True)  # Replace NaN with mean

# 3. One-Hot Encode the 'city' column
# Machine learning algorithms typically work with numbers, not text
# One-hot encoding converts categorical data (like city names) into binary columns
# Each unique city becomes its own column with 1 (true) or 0 (false) values
# Example: 'New York' becomes city_New York=1, city_London=0, city_Paris=0, etc.
cleaned_data = pd.get_dummies(cleaned_data, columns=['city'], prefix='city')
# columns=['city'] specifies which column to encode
# prefix='city' adds "city_" before each city name in the new column names

# Display the final cleaned and preprocessed data
print("\nCleaned and Preprocessed Data:")
print(cleaned_data)

# Key observations about the transformation:
# - Missing values are now filled in with reasonable estimates
# - The single 'city' column became multiple binary columns (one per unique city)
# - All data is now numerical, making it ready for machine learning algorithms
# - We went from 3 columns to 7 columns due to one-hot encoding
```

## **1.5 Using Data Transforms**

`scikit-learn` provides a `Pipeline` object to chain multiple processing steps together. This is extremely useful for organizing your workflow and preventing data leakage (applying information from the test set during training).

A pipeline encapsulates a sequence of transformers and a final estimator. When you call `fit()` on the pipeline, it applies `fit_transform()` for each transformer in sequence and then `fit()` on the final estimator.

```python
# Import necessary components for creating data processing pipelines
from sklearn.pipeline import make_pipeline  # Tool for chaining multiple processing steps together
from sklearn.impute import SimpleImputer    # Tool for filling in missing values automatically
from sklearn.cluster import KMeans          # Clustering algorithm (imported but not used in this example)

# Let's use the messy data from before, but only the numeric part
# We're selecting only the numerical columns because clustering algorithms work best with numbers
# The 'city' column contains text, so we'll exclude it for this demonstration
numeric_data = messy_data[['age', 'salary']]

# Create a pipeline that first imputes missing values, then scales the data
# A pipeline is like an assembly line - data flows through each step in order
# This is much cleaner than manually applying each transformation step by step
preprocessing_pipeline = make_pipeline(
    # Step 1: SimpleImputer fills in missing values (NaN) with the mean of each column
    # strategy='mean' tells it to replace missing values with the average of non-missing values
    # Alternative strategies include 'median', 'most_frequent', or 'constant'
    SimpleImputer(strategy='mean'),

    # Step 2: StandardScaler normalizes the data so all features have mean=0, std=1
    # This puts age (small numbers) and salary (large numbers) on the same scale
    # This step happens AFTER imputation, so the scaler sees complete data with no missing values
    StandardScaler()
)

# Fit and transform the data
# This is where the magic happens - the pipeline processes data through both steps automatically
# fit_transform() does two things:
# 1. fit(): learns the parameters (like mean values for imputation, mean/std for scaling)
# 2. transform(): applies the learned transformations to convert the data
transformed_data = preprocessing_pipeline.fit_transform(numeric_data)

# Display the original data with missing values
print("Original Numeric Data:")
print(numeric_data)

# Display the final processed data
# Notice how:
# - Missing values (NaN) have been replaced with calculated means
# - All values are now scaled to roughly the same range (typically -3 to +3)
# - The data is returned as a numpy array, not a DataFrame
print("\nData after Pipeline Transformation:")
print(transformed_data)

# Key advantages of using pipelines:
# 1. Cleaner code - no need to manually apply each step
# 2. Prevents data leakage - ensures proper order of operations
# 3. Easy to reuse - can apply the same pipeline to new data
# 4. Atomic operations - either all steps succeed or all fail
# 5. Great for machine learning workflows where you need consistent preprocessing
```

## **1.6 Dimensionality Reduction in Unsupervised Learning**

**Dimensionality reduction** aims to reduce the number of features in a dataset while retaining as much of the important information as possible.

**Principal Component Analysis (PCA)** is the most common technique. It transforms the original features into a new set of uncorrelated features called **principal components**. These components are ordered by the amount of variance they explain in the data. The first principal component captures the most variance, the second captures the next most, and so on.

By keeping only the first few principal components, we can reduce the dimensionality of the data with minimal loss of information.

```python
# Import necessary libraries for PCA (Principal Component Analysis) and sample data
from sklearn.decomposition import PCA  # Tool for reducing the number of features/dimensions in data
from sklearn.datasets import load_iris  # Function to load a classic, well-known dataset

# Load the iris dataset (a classic dataset for this purpose)
# The iris dataset contains measurements of 150 iris flowers across 4 features:
# - sepal length, sepal width, petal length, petal width
# It's perfect for demonstrating PCA because it has multiple related measurements
iris = load_iris()
X = iris.data    # The feature data (150 rows × 4 columns of measurements)
y = iris.target  # The labels (0, 1, 2 representing 3 different iris species)

# First, scale the data
# PCA is sensitive to the scale of features, so we must standardize first
# Without scaling, features with larger ranges would dominate the principal components
# For example, if one feature ranges 0-100 and another 0-1, PCA would focus on the 0-100 feature
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA to reduce to 2 components
# PCA finds the directions (principal components) that capture the most variance in the data
# n_components=2 means we want to reduce from 4 dimensions down to 2 dimensions
# This makes the data easier to visualize while preserving as much information as possible
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Show how the data shape changed
print(f"Original shape: {X_scaled.shape}")  # Should be (150, 4) - 150 samples, 4 features
print(f"Shape after PCA: {X_pca.shape}")   # Should be (150, 2) - 150 samples, 2 components

# Visualize the result
# Now we can plot our 4-dimensional data in a 2D scatter plot
plt.figure(figsize=(8, 6))
plt.title('PCA of Iris Dataset')

# Create a scatter plot where:
# - x-axis = first principal component (captures most variance)
# - y-axis = second principal component (captures second most variance)
# - colors represent the different iris species
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis', s=60)
plt.xlabel('First Principal Component')   # This axis explains the most variance
plt.ylabel('Second Principal Component')  # This axis explains the second most variance

# Add a legend showing which color represents which iris species
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
handles, labels = scatter.legend_elements()
plt.legend(handles=handles, labels=labels)  # labels will be ['0', '1', '2']
plt.show()

# Explained variance
# This tells us how much of the original information we preserved
# explained_variance_ratio_ shows what percentage of total variance each component captures
print(f"Explained variance by component: {pca.explained_variance_ratio_}")

# Calculate total information preserved by our 2 components
# If this number is high (like 95%+), we've successfully reduced dimensions without losing much info
# If it's low (like 60%), we might need more components or PCA might not be ideal for this data
print(f"Total variance explained by 2 components: {sum(pca.explained_variance_ratio_):.2%}")

# Key insights from this analysis:
# - We reduced 4-dimensional data to 2 dimensions for easy visualization
# - The scatter plot reveals natural clustering of the 3 iris species
# - The explained variance ratio tells us how much original information we retained
# - Higher explained variance = better dimensionality reduction
```

## **1.7 Clustering, k-Means Clustering**

**Clustering** is the process of partitioning data into groups (clusters) where data points in the same cluster are similar to each other.

**K-Means** is one of the most popular clustering algorithms. Its goal is to partition *n* observations into *k* clusters in which each observation belongs to the cluster with the nearest mean (cluster centroid).

**How K-Means Works:**
1.  **Initialize:** Randomly select *k* data points as initial centroids.
2.  **Assign:** Assign each data point to its closest centroid based on Euclidean distance.
3.  **Update:** Recalculate the centroid of each cluster by taking the mean of all data points assigned to it.
4.  **Repeat:** Repeat steps 2 and 3 until the centroids no longer move significantly or a maximum number of iterations is reached.

**Choosing 'k':** The number of clusters, *k*, must be specified beforehand. A common technique to find a good value for *k* is the **Elbow Method**, where we plot the sum of squared distances of samples to their closest cluster center for a range of *k* values. The 'elbow' point in the plot suggests an optimal *k*.

```python
from sklearn.cluster import KMeans  # Import the KMeans clustering algorithm
from sklearn.datasets import make_blobs  # Import a function to generate synthetic clustered data

# --- Generate synthetic data for clustering ---

# Create a dataset with 300 samples, 4 centers (clusters), and some variation (std=0.8)
# random_state=42 ensures reproducibility
X_blobs, y_blobs = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)

# --- Elbow Method to find optimal K ---

# Create an empty list to store the Within-Cluster Sum of Squares (WCSS) for each value of k
# WCSS tells us how tightly grouped the clusters are (lower is better)
wcss = []

# Try different values of k (number of clusters) from 1 to 10
for i in range(1, 11):
    # Create a KMeans model with:
    # - n_clusters=i: trying i clusters
    # - init='k-means++': smart initialization to improve results
    # - max_iter=300: max number of iterations for a single run
    # - n_init=10: run the algorithm 10 times with different initial centroids and pick the best result
    # - random_state=42: ensures consistent results
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)

    # Fit the model to the data
    kmeans.fit(X_blobs)

    # Append the WCSS (inertia) to the list for this value of k
    wcss.append(kmeans.inertia_)

# --- Plot the Elbow Curve ---

# Create a new figure to plot the WCSS values
plt.figure(figsize=(8, 5))

# Plot WCSS against the number of clusters (k)
# A sharp "elbow" in the plot indicates the optimal number of clusters
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')

# Add title and labels for clarity
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')

# Show grid lines for better readability
plt.grid(True)

# Display the elbow plot
plt.show()

# The elbow appears to be at k=4, which matches our synthetic data.
# --- Apply K-Means with k=4 ---

# Now we apply KMeans with the chosen value of k=4
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)

# Fit the model and get the predicted cluster labels for each data point
pred_y = kmeans.fit_predict(X_blobs)

# Get the coordinates of the final cluster centers
centroids = kmeans.cluster_centers_

# --- Visualize the Clusters ---

# Create a new figure for the cluster visualization
plt.figure(figsize=(8, 6))

# Plot the data points, colored by their predicted cluster
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=pred_y, s=50, cmap='viridis')

# Plot the cluster centroids as red 'X' marks
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Centroids')

# Add a title and axis labels
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show the legend to identify the centroids
plt.legend()

# Add grid lines for better visibility
plt.grid(True)

# Display the final cluster plot
plt.show()
```

## **1.8 Agglomerative Clustering**

**Agglomerative Clustering** is a type of **hierarchical clustering**. It follows a bottom-up approach:

1.  Start with each data point as its own cluster.
2.  Find the two most similar clusters and merge them.
3.  Repeat step 2 until only one cluster remains.

The similarity between clusters is determined by a **linkage criterion**:
-   **Ward:** Minimizes the variance of the clusters being merged.
-   **Complete (Maximum) Linkage:** Merges based on the maximum distance between points in the two clusters.
-   **Average Linkage:** Merges based on the average distance between all pairs of points in the two clusters.

A key advantage is that you don't need to specify the number of clusters beforehand. The entire hierarchy is built, which can be visualized using a **dendrogram**.

```python
from sklearn.cluster import AgglomerativeClustering  # Import Agglomerative Clustering model from scikit-learn
import scipy.cluster.hierarchy as sch                # Import SciPy's hierarchy tools to create dendrograms

# --- Plot the dendrogram to visualize the hierarchy ---

# Create a new figure with specific size for better clarity
plt.figure(figsize=(12, 7))

# Set the title of the plot to explain what it's showing
plt.title('Dendrogram')

# Create the dendrogram to visualize how clusters are formed
# sch.linkage builds the hierarchy using 'ward' linkage, which minimizes variance within clusters
# The dendrogram shows how data points are merged step-by-step into clusters
dend = sch.dendrogram(sch.linkage(X_blobs, method='ward'))

# Label the x-axis; each tick can represent a data point or a small cluster
plt.xlabel('Data Points')

# Label the y-axis to show the distance between merged clusters (dissimilarity)
plt.ylabel('Euclidean Distance')

# Display the dendrogram plot
plt.show()

# Based on the dendrogram, 4 clusters seems appropriate.
# --- Apply Agglomerative Clustering ---

# Create the Agglomerative Clustering model
# n_clusters=4 means we want to divide the data into 4 clusters
# metric='euclidean' defines the distance measure between points
# linkage='ward' minimizes the variance within clusters when merging
agg_cluster = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')

# Fit the model to the data and get the cluster labels for each point
# Each point is assigned to one of the 4 clusters
pred_y_agg = agg_cluster.fit_predict(X_blobs)

# --- Visualize the clusters ---

# Create a new figure for the cluster visualization
plt.figure(figsize=(8, 6))

# Plot the data points using a scatter plot
# X_blobs[:, 0] and X_blobs[:, 1] are the two features (x and y axes)
# c=pred_y_agg colors each point by its assigned cluster
# s=50 sets the point size
# cmap='plasma' sets the color map used for different clusters
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=pred_y_agg, s=50, cmap='plasma')

# Add a title and axis labels to the plot
plt.title('Agglomerative Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show grid lines to make the plot easier to read
plt.grid(True)

# Display the final clustering plot
plt.show()
```

## **1.9 DBSCAN**

**Density-Based Spatial Clustering of Applications with Noise (DBSCAN)** is a density-based clustering algorithm. Unlike K-Means, it can find arbitrarily shaped clusters and is robust to outliers (which it identifies as 'noise').

It works based on two key parameters:
-   **`eps` (epsilon):** The maximum distance between two samples for one to be considered as in the neighborhood of the other.
-   **`min_samples`:** The number of samples in a neighborhood for a point to be considered as a core point.

**How DBSCAN Works:**
1.  It starts with an arbitrary point that has not been visited.
2.  It retrieves the neighborhood of this point using `eps`.
3.  If there are enough neighbors (`min_samples`), a cluster is started. The point and its neighbors are added to the cluster.
4.  This process is expanded recursively for all neighbors.
5.  If a point does not have enough neighbors, it is labeled as noise (for now).
6.  The process continues until all points have been visited.

DBSCAN is powerful for datasets where clusters have irregular shapes and when you want to automatically detect noise.

```python
from sklearn.cluster import DBSCAN  # Import DBSCAN, a density-based clustering algorithm
from sklearn.datasets import make_moons  # Import a function to generate a dataset with crescent (moon-like) shapes

# --- Generate synthetic data (two interleaving half circles) ---

# Create a 2D dataset shaped like two moons
# n_samples=250 specifies the number of data points
# noise=0.1 adds some randomness to make the shapes more realistic (not perfectly clean)
# random_state=42 ensures consistent results each time the code is run
X_moons, y_moons = make_moons(n_samples=250, noise=0.1, random_state=42)

# Scale the data using StandardScaler so that all features have mean=0 and std=1
# This is important for DBSCAN, as it uses distance calculations
X_moons_scaled = StandardScaler().fit_transform(X_moons)

# --- Apply DBSCAN Clustering ---

# Create a DBSCAN model
# eps=0.3 defines the radius of the neighborhood around a point
# min_samples=5 means a core point must have at least 5 points (including itself) in its neighborhood
# These parameters help DBSCAN determine which points form dense regions (clusters)
dbscan = DBSCAN(eps=0.3, min_samples=5)

# Fit the DBSCAN model to the scaled data and predict the cluster labels
# DBSCAN assigns a label to each point:
# - A cluster number (e.g., 0, 1, 2, ...) for points in clusters
# - -1 for points that are considered noise (do not belong to any cluster)
pred_y_dbscan = dbscan.fit_predict(X_moons_scaled)

# --- Visualize the clustering results ---

# Create a new figure for the plot
plt.figure(figsize=(8, 6))

# Plot the data points
# Each point is colored based on its assigned cluster label
# c=pred_y_dbscan assigns colors to clusters; -1 (noise) will appear as a different color
# s=50 sets the size of each point
# cmap='viridis' sets the color palette
plt.scatter(X_moons_scaled[:, 0], X_moons_scaled[:, 1], c=pred_y_dbscan, s=50, cmap='viridis')

# Add a title and axis labels to explain the plot
plt.title('DBSCAN Clustering on Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Note: DBSCAN uses the label -1 to mark noise points (those not assigned to any cluster)

# Calculate the number of clusters by counting unique labels
# Subtract 1 if -1 (noise) is present, because it's not a real cluster
n_clusters_ = len(set(pred_y_dbscan)) - (1 if -1 in pred_y_dbscan else 0)

# Print the estimated number of clusters found
print(f"Estimated number of clusters: {n_clusters_}")

# Add grid lines to the plot for better readability
plt.grid(True)

# Display the final plot
plt.show()
```
